```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
ax.plot(x,y)
display(fig)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

display(fig)
```

```python
ax1.plot(x, y, color='red')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x, y, color='red')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
display(fig)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
display(fig)
```

```python
ax1.plot(x, z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
display(fig)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 2))

axes[0].plot(x, y, color='blue', lw=5, ls='--')
axes[1].plot(x, z, color='red', lw=3)
display(fig)
```
