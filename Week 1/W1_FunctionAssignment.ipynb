```python
#ANIMAL CRACKERS

def animal_crackers(text):
    words = text.split()
    word1 = words[0]
    word2 = words[1]
    
    return word1[0].lower() == word2[0].lower()
```

```python
# MAKES TWENTY

def makes_twenty(n1, n2):
    return (n1 + n2) == 20 or n1 == 20 or n2 == 20
```

```python
#Level 1

#Old MACDONALD

def old_macdonald(name):
    first_part = name[:3]
    second_part = name[3:]
    
    return first_part.capitalize() + second_part.capitalize()
```

```python
#MASTER YODA

def master_yoda(text):
    words = text.split()
    words.reverse()
    return " ".join(words)
```

```python
#ALMOST THERE

def almost_there(n):
    check1 = abs(100 - n) <= 10
    check2 = abs(200 - n) <= 10
    
    return check1 or check2
```

```python
#Find 30

def has_33(nums):
    for i in range(len(nums) - 1):
        if nums[i] == 3 and nums[i+1] == 3:
            return True
    return False

```

```python
#Paper Doll

def paper_doll(text):
    answer = ""
    for letter in text:
        answer = answer + letter * 3
    return answer
```

```python
#BlackJack

def blackjack(a, b, c):
    total = a + b + c
    
    if total <= 21:
        return total
        
    if 11 in [a, b, c] and total > 21:
        total = total - 10
        
    if total > 21:
        return 'BUST'
        
    return total
```

```python
#Summer of 69

def summer_69(arr):
    total = 0
    add = True
    
    for num in arr:
        if add:
            if num == 6:
                add = False
            else:
                total += num
        else:
            if num == 9:
                add = True
                
    return total
```

```python
#SPY GAME

def spy_game(nums):
    target = [0, 0, 7]
    index = 0
    
    for num in nums:
        if num == target[index]:
            index += 1
            if index == 3:
                return True
    return False
```

```python
#Count primes

def check_prime(num):
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

def count_primes(num):
    count = 0
    for n in range(2, num + 1):
        if check_prime(n):
            count += 1
    return count
```
