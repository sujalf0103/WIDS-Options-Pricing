```python
import random

target = random.randint(1, 100)
guesses = [0]

while True:
    guess = int(input("Guess a number between 1 and 100: "))
    
    if guess < 1 or guess > 100:
        print('OUT OF BOUNDS')
        continue
        
    if guess == target:
        print(f'CONGRATS! YOU GUESSED IT IN {len(guesses)} GUESSES')
        break
        
    guesses.append(guess)
    
    if len(guesses) == 2:
        if abs(target - guess) <= 10:
            print('WARM!')
        else:
            print('COLD!')
    else:
        if abs(target - guess) < abs(target - guesses[-2]):
            print('WARMER!')
        else:
            print('COLDER!')
```
